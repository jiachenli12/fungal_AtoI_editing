```python
def read_fasta(file_path):
    sequences = {}
    with open(file_path, 'r') as f:
        seq_id = None
        seq_lines = []
        for line in f:
            line = line.strip()
            if line.startswith('>'):
                if seq_id:
                    sequences[seq_id] = ''.join(seq_lines)
                seq_id = line[1:]
                seq_lines = []
            else:
                seq_lines.append(line)
        if seq_id:
            sequences[seq_id] = ''.join(seq_lines)
    return sequences
```

```python
def add_edit_pos_in_cds(gff_file, edit_data_df, chrom_map):
    # Load GFF file
    gff_df = pd.read_csv(gff_file, sep='\t', comment='#', header=None,
                         names=["seqid", "source", "type", "start", "end", "score", "strand", "phase", "attributes"])
    
    # Initialize gene and transcript mappings
    gene_ids = set()
    transcript_to_gene = {}
    gene_to_transcript = {}

    # First pass: collect gene IDs and build transcript-gene mappings
    for _, row in gff_df.iterrows():
        attr = row["attributes"]
        if row["type"] == "gene":
            if match := re.search(r'ID=([^;]+)', attr):
                gene_id = match.group(1)
                gene_ids.add(gene_id)
        elif row["type"] in {"mRNA", "transcript"}:
            gene_match = re.search(r'Parent=([^;]+)', attr)
            tx_match = re.search(r'ID=([^;]+)', attr)
            if gene_match and tx_match:
                gene_id = gene_match.group(1)
                tx_id = tx_match.group(1)
                transcript_to_gene[tx_id] = gene_id
                gene_to_transcript[gene_id] = tx_id
                gene_ids.add(gene_id)

    # Extract CDS features
    cds_df = gff_df[gff_df["type"] == "CDS"].copy()
    
    # Extract parent from attributes
    cds_df["parent"] = cds_df["attributes"].str.extract(r'Parent=([^;]+)')
    
    # Map to gene ID
    def map_to_gene_id(parent):
        if parent in gene_ids:
            return parent  # Direct gene parent
        if parent in transcript_to_gene:
            return transcript_to_gene[parent]  # Transcript parent
        return None
    
    cds_df["gene_id"] = cds_df["parent"].apply(map_to_gene_id)
    cds_df = cds_df.dropna(subset=["gene_id"])

    # Build gene → CDS dictionary
    gene_cds_dict = defaultdict(lambda: {"strand": None, "cds": []})
    for _, row in cds_df.iterrows():
        gene_id = row["gene_id"]
        if gene_cds_dict[gene_id]["strand"] is None:
            gene_cds_dict[gene_id]["strand"] = row["strand"]
        gene_cds_dict[gene_id]["cds"].append({
            "seqid": row["seqid"],
            "start": int(row["start"]),
            "end": int(row["end"])
        })

    # Helper to extract chrom and coordinate
    def extract_chrom_coord(coord_str):
        chrom, pos = coord_str.split(":")
        return chrom, int(pos)

    # Core logic to compute CDS position
    def get_edit_pos_in_cds(gene_id, chrom, edit_pos):
        entry = gene_cds_dict.get(gene_id)
        if entry is None:
            return None

        cds_list = [cds for cds in entry["cds"] if cds["seqid"] == chrom]
        if not cds_list:
            return None

        strand = entry["strand"]
        if strand == '+':
            cds_list.sort(key=lambda x: x["start"])
        else:
            cds_list.sort(key=lambda x: x["end"], reverse=True)

        cds_len = 0
        for cds in cds_list:
            if cds["start"] <= edit_pos <= cds["end"]:
                if strand == '+':
                    return cds_len + (edit_pos - cds["start"] + 1)
                else:
                    return cds_len + (cds["end"] - edit_pos + 1)
            cds_len += cds["end"] - cds["start"] + 1
        return None

    # Compute CDS position for each edit
    def compute_cds_pos(row):
        chrom_short, edit_pos = extract_chrom_coord(row['Coordinate'])
        chrom_full = chrom_map.get(chrom_short, chrom_short).split()[0]
        
        # EXTRACT ONLY GENE ID (REMOVE PARENTHESES CONTENT)
        raw_gene_id = row['GeneID(Symbol)']
        clean_gene_id = raw_gene_id.split()[0]  # Takes "FGRAMPH1_01T00243" from "FGRAMPH1_01T00243 (nat1)"
        gene_id = "gene-" + clean_gene_id
        
        return get_edit_pos_in_cds(gene_id, chrom_full, edit_pos)

    # Apply and return modified DataFrame
    edit_data_df = edit_data_df.copy()
    edit_data_df['edit_pos_in_cds'] = edit_data_df.apply(compute_cds_pos, axis=1)
    return edit_data_df
```

```python
def is_stop_to_or_from_nonsense(aa_change):
    if not isinstance(aa_change, str):
        return False
    parts = aa_change.split('/')
    if len(parts) != 2:
        return False
    ref, edited = parts
    return ('*' in ref) ^ ('*' in edited)  

nc_mask = nc_edit_data_TA_sub_non['Amino Acid'].apply(is_stop_to_or_from_nonsense)
nt_mask = nt_edit_data_TA_sub_non['Amino Acid'].apply(is_stop_to_or_from_nonsense)
sm_mask = sm_edit_data_TA_sub_non['Amino Acid'].apply(is_stop_to_or_from_nonsense)
po_mask = po_edit_data_TA_sub_non['Amino Acid'].apply(is_stop_to_or_from_nonsense)
fg_mask = fg_edit_data_TA_sub_non['Amino Acid'].apply(is_stop_to_or_from_nonsense)

nc_edit_data_TA_sub_non_filtered = nc_edit_data_TA_sub_non[~nc_mask].copy()
nt_edit_data_TA_sub_non_filtered = nt_edit_data_TA_sub_non[~nt_mask].copy()
sm_edit_data_TA_sub_non_filtered = sm_edit_data_TA_sub_non[~sm_mask].copy()
po_edit_data_TA_sub_non_filtered = po_edit_data_TA_sub_non[~po_mask].copy()
fg_edit_data_TA_sub_non_filtered = fg_edit_data_TA_sub_non[~fg_mask].copy()
```

```python
species_data = [
    (species_labels[sp], fractions[sp]['fs'], fractions[sp]['fd'], fractions[sp]['fr'], pvals_div[sp], pvals_rest[sp])
    for sp in species_order
]

# Parameters
bar_width = 0.5
group_spacing = 0.4  # space between species
bar_spacing = 0   # space between bars within a group

# Set up figure
fig, ax = plt.subplots(figsize=(12, 6))

# X-tick locations and labels
xticks = []
xlabels = []

# Colors
colors = ['grey', 'red', 'dodgerblue']
labels = [
    'Synonymous (${\it F}_{\mathrm{S}}$)',  
    'Diversifying (${\it F}_{\mathrm{R}}$)',  
    'Restorative (${\it F}_{\mathrm{D}}$)'  
]

# Loop through species and plot bars
for i, (label, fs, fd, fr, pval_d, pval_r) in enumerate(species_data):
    base_x = i * (3 * bar_width + group_spacing)
    bar_xs = [base_x + j * (bar_width + bar_spacing) for j in range(3)]
    values = [fs, fd, fr]

    # Plot bars
    for x, val, color in zip(bar_xs, values, colors):
        ax.bar(x, val, color=color, width=bar_width)

    # Add significance stars above Diversifying (1) and Restorative (2)
    stars_d = "***" if pval_d < 0.001 else "**" if pval_d < 0.01 else "*" if pval_d < 0.05 else ""
    stars_r = "***" if pval_r < 0.001 else "**" if pval_r < 0.01 else "*" if pval_r < 0.05 else ""

    if stars_d:
        ax.text(bar_xs[1], values[1] + 0.003, stars_d, ha='center', fontsize=16)
    if stars_r:
        ax.text(bar_xs[2], values[2] + 0.003, stars_r, ha='center', fontsize=16)

    # Store tick position and label
    xticks.append(base_x + bar_width + bar_spacing)
    xlabels.append(label)

# Set ticks and labels
ax.set_xticks(xticks)
ax.set_xticklabels(xlabels, fontsize=16)

# Y-axis
ax.set_ylabel("Fraction of sites edited", fontsize=18)
ax.tick_params(axis='y', labelsize=14)

# Legend
handles = [plt.Rectangle((0, 0), 1, 1, color=c, label=l) for c, l in zip(colors, labels)]
ax.legend(handles=handles, fontsize=14, frameon=False, loc='upper right', ncol=1)

# Style
sns.despine()
plt.tight_layout()
plt.show()
```

```python
species_data = [
    (species_labels[sp], fractions[sp]['fs'], fractions[sp]['fd'], fractions[sp]['fr'], pvals_div[sp], pvals_rest[sp])
    for sp in species_order
]

# Parameters
bar_width = 0.5
group_spacing = 0.4  # space between species
bar_spacing = 0   # space between bars within a group

# Set up figure
fig, ax = plt.subplots(figsize=(12, 6))

# X-tick locations and labels
xticks = []
xlabels = []

# Colors
colors = ['grey', 'red', 'dodgerblue']
labels = [
    'Synonymous (${\it F}_{\mathrm{S}}$)',  
    'Diversifying (${\it F}_{\mathrm{R}}$)',  
    'Restorative (${\it F}_{\mathrm{D}}$)'  
]

# Loop through species and plot bars
for i, (label, fs, fd, fr, pval_d, pval_r) in enumerate(species_data):
    base_x = i * (3 * bar_width + group_spacing)
    bar_xs = [base_x + j * (bar_width + bar_spacing) for j in range(3)]
    values = [fs, fd, fr]

    # Plot bars
    for x, val, color in zip(bar_xs, values, colors):
        ax.bar(x, val, color=color, width=bar_width)

    # Add significance stars above Diversifying (1) and Restorative (2)
    stars_d = "***" if pval_d < 0.001 else "**" if pval_d < 0.01 else "*" if pval_d < 0.05 else ""
    stars_r = "***" if pval_r < 0.001 else "**" if pval_r < 0.01 else "*" if pval_r < 0.05 else ""

    if stars_d:
        ax.text(bar_xs[1], values[1] + 0.003, stars_d, ha='center', fontsize=16)
    if stars_r:
        ax.text(bar_xs[2], values[2] + 0.003, stars_r, ha='center', fontsize=16)

    # Store tick position and label
    xticks.append(base_x + bar_width + bar_spacing)
    xlabels.append(label)

# Set ticks and labels
ax.set_xticks(xticks)
ax.set_xticklabels(xlabels, fontsize=16)

# Y-axis
ax.set_ylabel("Fraction of sites edited at > 10% level", fontsize=18)
ax.tick_params(axis='y', labelsize=14)

# Legend
handles = [plt.Rectangle((0, 0), 1, 1, color=c, label=l) for c, l in zip(colors, labels)]
ax.legend(handles=handles, fontsize=14, frameon=False, loc='upper right', ncol=1)

# Style
sns.despine()
plt.tight_layout()
plt.show()
```

```python
box_colors = ['grey', 'dodgerblue', 'red']  # Syn, Rest, Div
edit_types = ['Synonymous', 'Restorative', 'Diversifying']

species_labels = {
    'fg': r"$\it{F.\ graminearum}$",
    'nc': r"$\it{N.\ crassa}$",
    'nt': r"$\it{N.\ tetrasperma}$",
    'sm': r"$\it{S.\ macrospora}$",
    'po': r"$\it{P.\ omphalodes}$"
}
species_order = ['fg', 'nc', 'nt', 'sm', 'po']

# Collect data - order: [syn, rest, div]
all_data = []
for sp in species_order:
    try:
        s_edit = eval(f"{sp}_s_editingLevel").dropna() * 100
        r_edit = eval(f"{sp}_r_editingLevel").dropna() * 100
        d_edit = eval(f"{sp}_d_editingLevel").dropna() * 100
    except:
        s_edit = r_edit = d_edit = pd.Series(dtype=float)
    all_data.append([s_edit, r_edit, d_edit])

# Plot setup
fig, ax = plt.subplots(figsize=(13, 6))
positions = []
box_data = []
colors = []
xticks = []
xtick_labels = []

spacing = 4
width = 0.6

for i, (sp, (s, r, d)) in enumerate(zip(species_order, all_data)):
    base = i * spacing
    box_data.extend([s, r, d])
    positions.extend([base, base + 1, base + 2])
    colors.extend(box_colors)
    xticks.append(base + 1)
    xtick_labels.append(species_labels[sp])

# Create boxplots
bp = ax.boxplot(
    box_data,
    positions=positions,
    widths=width,
    patch_artist=True,
    medianprops=dict(color='black', linewidth=1.5),
    showfliers=False
)

# Color boxes
for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_linewidth(0)

# Set axes
ax.set_ylabel("Editing Level (%)", fontsize=16)
ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels, fontsize=14)
ax.tick_params(axis='y', labelsize=14)

# Significance stars
def get_stars(pval):
    if pval < 0.001:
        return '***'
    elif pval < 0.01:
        return '**'
    elif pval < 0.05:
        return '*'
    else:
        return 'ns'

for i, (s, r, d) in enumerate(all_data):
    base = i * spacing

    # Guard for empty sets
    if len(s) == 0 or len(r) == 0 or len(d) == 0:
        continue

    y_max = np.nanmax([s.max(), r.max(), d.max()]) + 1.5

    # Diversifying vs Synonymous 
    try:
        pval_d = mannwhitneyu(d, s, alternative='two-sided').pvalue
        stars_d = get_stars(pval_d)
        line_y = y_max + 10
        text_y = line_y + (0.8 if stars_d != 'ns' else 1.5)  # Higher for ns
        ax.plot([base, base + 2], [line_y, line_y], color='black', linewidth=1)
        ax.text(base + 1, text_y, stars_d, ha='center', fontsize=14)
    except:
        pass

    # Restorative vs Synonymous 
    try:
        pval_r = mannwhitneyu(r, s, alternative='two-sided').pvalue
        stars_r = get_stars(pval_r)
        line_y = y_max
        text_y = line_y + (0.8 if stars_r != 'ns' else 1.5)  # Higher for ns
        ax.plot([base, base + 1], [line_y, line_y], color='black', linewidth=1)
        ax.text(base + 0.5, text_y, stars_r, ha='center', fontsize=14)
    except:
        pass

# Legend
handles = [
    plt.Rectangle((0, 0), 1, 1, color=box_colors[0], label='Synonymous (${\it L}_{\mathrm{S}}$)'),
    plt.Rectangle((0, 0), 1, 1, color=box_colors[1], label='Restorative (${\it L}_{\mathrm{R}}$)'),
    plt.Rectangle((0, 0), 1, 1, color=box_colors[2], label='Diversifying (${\it L}_{\mathrm{D}}$)'),
]
ax.legend(handles=handles, fontsize=12, frameon=False, loc='upper right')

sns.despine()
plt.tight_layout()
# plt.savefig('editing_level_boxplot_fixed.pdf', dpi=300)
plt.show()
```
