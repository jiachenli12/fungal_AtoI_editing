```python
def merge_edits(edit_df, potential_df, effect_type, kind):
    filtered = edit_df[edit_df['Effect'] == effect_type]

    merged = filtered.merge(
        potential_df,
        left_on=['GeneID(Symbol)', 'edit_pos_in_cds'],
        right_on=['gene', 'gap_adjusted_position'],
        how='inner',
        suffixes=('_edit', f'_{kind}')
    )

    if 'base_before_edit_edit' in merged.columns:
        merged['base_before_edit'] = merged['base_before_edit_edit']
    elif f'base_before_edit_{kind}' in merged.columns:
        merged['base_before_edit'] = merged[f'base_before_edit_{kind}']

    cols = [
        'GeneID(Symbol)', 'species', 'Region', 'Position', 'Coverage-q30',
        'Codon', 'AA', 'Effect',
        'edit_pos_in_cds', 'base_before_edit', 'gap_adjusted_position',
        'ref_codon', 'mutated_codon', 'ref_aa', 'mutated_aa', 'node_aa',
        'editing_level'
    ]

    cols_present = [c for c in cols if c in merged.columns]
    return merged[cols_present]

fg_div = merge_edits(fg_editing_sites_nonsyn, fg_potential_diversifying, 'non-syn', 'div')
fg_rest = merge_edits(fg_editing_sites_nonsyn, fg_potential_restorative, 'non-syn', 'rest')
fg_syn = merge_edits(fg_editing_sites_syn, fg_potential_syn, 'syn', 'syn')

nc_div = merge_edits(nc_editing_sites_nonsyn, nc_potential_diversifying, 'non-syn', 'div')
nc_rest = merge_edits(nc_editing_sites_nonsyn, nc_potential_restorative, 'non-syn', 'rest')
nc_syn = merge_edits(nc_editing_sites_syn, nc_potential_syn, 'syn', 'syn')

nt_div = merge_edits(nt_editing_sites_nonsyn, nt_potential_diversifying, 'non-syn', 'div')
nt_rest = merge_edits(nt_editing_sites_nonsyn, nt_potential_restorative, 'non-syn', 'rest')
nt_syn = merge_edits(nt_editing_sites_syn, nt_potential_syn, 'syn', 'syn')
```

```python
MOTIFS = ["TA", "AA", "CA", "GA"]

def count_by_motif(df, motifs=MOTIFS, base_col="base_before_edit"):
    out = {m: 0 for m in motifs}
    if df is None or len(df) == 0:
        return out
    motif_series = (df[base_col].astype(str).str.upper() + "A")
    vc = motif_series.value_counts()
    for m in motifs:
        if m in vc:
            out[m] = int(vc[m])
    return out

def compute_RDS_for_species(
    rest_pot_df,
    div_pot_df,
    syn_pot_df,
    motifs=MOTIFS
):
    R_pot = count_by_motif(rest_pot_df, motifs=motifs)
    D_pot = count_by_motif(div_pot_df, motifs=motifs)
    S_pot = count_by_motif(syn_pot_df, motifs=motifs)
    restorative_total = sum(R_pot[m] for m in motifs)
    diversifying_total = sum(D_pot[m] for m in motifs)
    syn_total = sum(S_pot[m] for m in motifs)
    motif_counts = {m: R_pot[m] + D_pot[m] + S_pot[m] for m in motifs}
    return {
        "syn_total": syn_total,
        "restorative_total": restorative_total,
        "diversifying_total": diversifying_total,
        "motif_counts": motif_counts,
        "syn_by_motif": S_pot,
        "restorative_by_motif": R_pot,
        "diversifying_by_motif": D_pot,
    }

fg_row = {"species": "Fg", **compute_RDS_for_species(
    rest_pot_df = fg_potential_restorative,
    div_pot_df  = fg_potential_diversifying,
    syn_pot_df  = fg_potential_syn,
)}

nc_row = {"species": "Nc", **compute_RDS_for_species(
    rest_pot_df = nc_potential_restorative,
    div_pot_df  = nc_potential_diversifying,
    syn_pot_df  = nc_potential_syn,
)}

nt_row = {"species": "Nt", **compute_RDS_for_species(
    rest_pot_df = nt_potential_restorative,
    div_pot_df  = nt_potential_diversifying,
    syn_pot_df  = nt_potential_syn,
)}

summary_df = pd.DataFrame([fg_row, nc_row, nt_row])
```

```python
BASES = ["T", "A", "C", "G"]

species_dfs = {
    "fg": {"div": fg_div, "rest": fg_rest, "syn": fg_syn},
    "nc": {"div": nc_div, "rest": nc_rest, "syn": nc_syn},
    "nt": {"div": nt_div, "rest": nt_rest, "syn": nt_syn},
}

for sp, classes in species_dfs.items():
    for cls, df in classes.items():
        for b in BASES:
            var_name = f"{sp}_{cls}_{b}A"  
            globals()[var_name] = df[df["base_before_edit"] == b].copy()
```

```python
fg_prob_TA_edited = (len(fg_div_TA)+len(fg_rest_TA)+len(fg_syn_TA))/ (summary_df['motif_counts'][0]['TA'])
fg_prob_CA_edited = (len(fg_div_CA)+len(fg_rest_CA)+len(fg_syn_CA))/ (summary_df['motif_counts'][0]['CA'])
fg_prob_AA_edited = (len(fg_div_AA)+len(fg_rest_AA)+len(fg_syn_AA))/ (summary_df['motif_counts'][0]['AA'])
fg_prob_GA_edited = (len(fg_div_GA)+len(fg_rest_GA)+len(fg_syn_GA))/ (summary_df['motif_counts'][0]['GA'])

nc_prob_TA_edited = (len(nc_div_TA)+len(nc_rest_TA)+len(nc_syn_TA))/ (summary_df['motif_counts'][1]['TA'])
nc_prob_CA_edited = (len(nc_div_CA)+len(nc_rest_CA)+len(nc_syn_CA))/ (summary_df['motif_counts'][1]['CA'])
nc_prob_AA_edited = (len(nc_div_AA)+len(nc_rest_AA)+len(nc_syn_AA))/ (summary_df['motif_counts'][1]['AA'])
nc_prob_GA_edited = (len(nc_div_GA)+len(nc_rest_GA)+len(nc_syn_GA))/ (summary_df['motif_counts'][1]['GA'])

nt_prob_TA_edited = (len(nt_div_TA)+len(nt_rest_TA)+len(nt_syn_TA))/ (summary_df['motif_counts'][2]['TA'])
nt_prob_CA_edited = (len(nt_div_CA)+len(nt_rest_CA)+len(nt_syn_CA))/ (summary_df['motif_counts'][2]['CA'])
nt_prob_AA_edited = (len(nt_div_AA)+len(nt_rest_AA)+len(nt_syn_AA))/ (summary_df['motif_counts'][2]['AA'])
nt_prob_GA_edited = (len(nt_div_GA)+len(nt_rest_GA)+len(nt_syn_GA))/ (summary_df['motif_counts'][2]['GA'])
```

```python
MOTIFS = ["TA", "AA", "CA", "GA"]
BASES = ["T", "A", "C", "G"]

species_dfs = {
    "Fg": {"div": fg_div, "rest": fg_rest, "syn": fg_syn},
    "Nc": {"div": nc_div, "rest": nc_rest, "syn": nc_syn},
    "Nt": {"div": nt_div, "rest": nt_rest, "syn": nt_syn},
}

def counts_by_base(df, base_col="base_before_edit"):
    vc = df[base_col].astype(str).str.upper().value_counts()
    return {b: int(vc.get(b, 0)) for b in BASES}

def edited_probs_for_species(sp, row):
    div_df, rest_df, syn_df = species_dfs[sp]["div"], species_dfs[sp]["rest"], species_dfs[sp]["syn"]
    edited_counts = {b: 0 for b in BASES}
    for df in (div_df, rest_df, syn_df):
        cb = counts_by_base(df)
        for b in BASES:
            edited_counts[b] += cb[b]
    motif_counts = row["motif_counts"]
    probs = {}
    for b in BASES:
        mot = b + "A"
        den = int(motif_counts.get(mot, 0))
        num = edited_counts[b]
        probs[mot] = (num / den) if den else 0.0
    return probs

expected_syn = []
expected_rest = []
expected_div = []

for i, row in summary_df.iterrows():
    sp = row["species"]
    probs = edited_probs_for_species(sp, row)
    S = sum(row["syn_by_motif"][m] * probs[m] for m in MOTIFS)
    R = sum(row["restorative_by_motif"][m] * probs[m] for m in MOTIFS)
    D = sum(row["diversifying_by_motif"][m] * probs[m] for m in MOTIFS)
    expected_syn.append(S)
    expected_rest.append(R)
    expected_div.append(D)

summary_df["weighted_S"] = expected_syn
summary_df["weighted_R"] = expected_rest
summary_df["weighted_D"] = expected_div
```

```python
denoms = {
    'Fg': summary_df.loc[summary_df['species']=='Fg', 'restorative_by_motif'].iloc[0],
    'Nc': summary_df.loc[summary_df['species']=='Nc', 'restorative_by_motif'].iloc[0],
    'Nt': summary_df.loc[summary_df['species']=='Nt', 'restorative_by_motif'].iloc[0],
}

div_denoms = {
    'Fg': summary_df.loc[summary_df['species']=='Fg', 'diversifying_by_motif'].iloc[0],
    'Nc': summary_df.loc[summary_df['species']=='Nc', 'diversifying_by_motif'].iloc[0],
    'Nt': summary_df.loc[summary_df['species']=='Nt', 'diversifying_by_motif'].iloc[0],
}

syn_denoms = {
    'Fg': summary_df.loc[summary_df['species']=='Fg', 'syn_by_motif'].iloc[0],
    'Nc': summary_df.loc[summary_df['species']=='Nc', 'syn_by_motif'].iloc[0],
    'Nt': summary_df.loc[summary_df['species']=='Nt', 'syn_by_motif'].iloc[0],
}

for sp in ['Fg', 'Nc', 'Nt']:
    print(f'\n {sp} ')
    for motif in ['GA', 'CA', 'AA', 'TA']:
        rest = len(eval(f"{sp.lower()}_rest_{motif}")) / denoms[sp][motif]
        div  = len(eval(f"{sp.lower()}_div_{motif}"))  / div_denoms[sp][motif]
        syn  = len(eval(f"{sp.lower()}_syn_{motif}"))  / syn_denoms[sp][motif]
        print(f'{motif}:  Rest={rest:.4f},  Div={div:.4f},  Syn={syn:.4f}')
```

```python
fg_total_A = int(summary_df.loc[summary_df["species"]=="Fg", ["syn_total","restorative_total","diversifying_total"]].sum(axis=1).iloc[0])
nc_total_A = int(summary_df.loc[summary_df["species"]=="Nc", ["syn_total","restorative_total","diversifying_total"]].sum(axis=1).iloc[0])
nt_total_A = int(summary_df.loc[summary_df["species"]=="Nt", ["syn_total","restorative_total","diversifying_total"]].sum(axis=1).iloc[0])

_weights = (
    summary_df.set_index("species")[["weighted_S","weighted_R","weighted_D"]]
    .rename(columns={"weighted_S":"S","weighted_R":"R","weighted_D":"D"})
    .to_dict(orient="index")
)

_species_ctx = {
    "Fg": {"total_A": float(fg_total_A), "syn_len": int(len(fg_syn)), "rest_len": int(len(fg_rest)), "div_len": int(len(fg_div))},
    "Nc": {"total_A": float(nc_total_A), "syn_len": int(len(nc_syn)), "rest_len": int(len(nc_rest)), "div_len": int(len(nc_div))},
    "Nt": {"total_A": float(nt_total_A), "syn_len": int(len(nt_syn)), "rest_len": int(len(nt_rest)), "div_len": int(len(nt_div))},
}

def motif_weighted_denominators(S, R, D, total_A):
    denom_sum = (S + R + D)
    return {"S": total_A * (S / denom_sum),
            "R": total_A * (R / denom_sum),
            "D": total_A * (D / denom_sum)}

rows = []
for sp in ["Fg", "Nc", "Nt"]:
    S_w, R_w, D_w = _weights[sp]["S"], _weights[sp]["R"], _weights[sp]["D"]
    totA = _species_ctx[sp]["total_A"]
    syn_len = _species_ctx[sp]["syn_len"]
    rest_len = _species_ctx[sp]["rest_len"]
    div_len = _species_ctx[sp]["div_len"]

    denoms = motif_weighted_denominators(S_w, R_w, D_w, totA)
    denom_S, denom_R, denom_D = float(denoms["S"]), float(denoms["R"]), float(denoms["D"])

    Fs_w = syn_len  / denom_S if denom_S > 0 else 0.0
    Fr_w = rest_len / denom_R if denom_R > 0 else 0.0
    Fd_w = div_len  / denom_D if denom_D > 0 else 0.0

    table_r_s = np.array([[rest_len, max(denom_R - rest_len, 0.0)],
                          [syn_len,  max(denom_S - syn_len,  0.0)]], dtype=float)
    table_d_s = np.array([[div_len,  max(denom_D - div_len,  0.0)],
                          [syn_len,  max(denom_S - syn_len,  0.0)]], dtype=float)

    _, p_r_s, _, _ = chi2_contingency(table_r_s, correction=False)
    _, p_d_s, _, _ = chi2_contingency(table_d_s, correction=False)

    rows.append({"species": sp,
                 "Fs_weighted": Fs_w, "Fr_weighted": Fr_w, "Fd_weighted": Fd_w,
                 "p_r_vs_s": p_r_s, "p_d_vs_s": p_d_s})

chi2_results_df = pd.DataFrame(rows)

summary_df = summary_df.copy()
for sp in ["Fg","Nc","Nt"]:
    m = summary_df["species"].eq(sp)
    summary_df.loc[m, "Fs_weighted"] = float(chi2_results_df.loc[chi2_results_df["species"].eq(sp), "Fs_weighted"].iloc[0])
    summary_df.loc[m, "Fr_weighted"] = float(chi2_results_df.loc[chi2_results_df["species"].eq(sp), "Fr_weighted"].iloc[0])
    summary_df.loc[m, "Fd_weighted"] = float(chi2_results_df.loc[chi2_results_df["species"].eq(sp), "Fd_weighted"].iloc[0])


species_order = ["Fg","Nc","Nt"]
species_labels = {
    "Fg": r"$\it{F.\ graminearum}$",
    "Nc": r"$\it{N.\ crassa}$",
    "Nt": r"$\it{N.\ tetrasperma}$",
}

plot_df = summary_df.set_index("species").loc[species_order].reset_index()
plot_df = plot_df.merge(chi2_results_df[["species","p_r_vs_s","p_d_vs_s"]], on="species", how="left")

def sig_symbol(p):
    if pd.isna(p): return "na"
    if p < 1e-4:   return "****"
    if p < 1e-3:   return "***"
    if p < 1e-2:   return "**"
    if p < 5e-2:   return "*"
    return "ns"

plot_df["sig_r_vs_s"] = plot_df["p_r_vs_s"].apply(sig_symbol)
plot_df["sig_d_vs_s"] = plot_df["p_d_vs_s"].apply(sig_symbol)

Fs_vals = plot_df["Fs_weighted"].values.astype(float)
Fr_vals = plot_df["Fr_weighted"].values.astype(float)
Fd_vals = plot_df["Fd_weighted"].values.astype(float)

x = np.arange(len(species_order))
bar_width = 0.25
x_fs, x_fr, x_fd = x - bar_width, x, x + bar_width

fig, ax = plt.subplots(figsize=(13,6))
ax.bar(x_fs, Fs_vals, width=bar_width, color="grey",       label=r"Synonymous")
ax.bar(x_fr, Fr_vals, width=bar_width, color="dodgerblue", label=r"Restorative")
ax.bar(x_fd, Fd_vals, width=bar_width, color="red",        label=r"Diversifying")

ax.set_xticks(x)
ax.set_xticklabels([species_labels[s] for s in species_order], fontsize=22)
ax.set_ylabel("Fraction of sites edited", fontsize=22)
ax.tick_params(axis="y", labelsize=22)
sns.despine()
ax.legend(frameon=False, fontsize=22, loc="upper center", bbox_to_anchor=(0.5, -0.12), ncol=3)

for i in range(len(species_order)):
    y_max = max(Fs_vals[i], Fr_vals[i], Fd_vals[i])
    off = 0.01 * (y_max if y_max > 0 else 1.0)
    ax.text(x_fr[i], Fr_vals[i] + off, plot_df.loc[i, "sig_r_vs_s"], ha="center", va="bottom", fontsize=22)
    ax.text(x_fd[i], Fd_vals[i] + off, plot_df.loc[i, "sig_d_vs_s"], ha="center", va="bottom", fontsize=22)

plt.tight_layout()
plt.show()

print(summary_df[["species","Fs_weighted","Fr_weighted","Fd_weighted"]])
```

```python
def to_series(df):
    return pd.to_numeric(df['editing_level'], errors='coerce').dropna().astype(float)

fg_s_editingLevel = to_series(fg_syn)
fg_r_editingLevel = to_series(fg_rest)
fg_d_editingLevel = to_series(fg_div)

nc_s_editingLevel = to_series(nc_syn)
nc_r_editingLevel = to_series(nc_rest)
nc_d_editingLevel = to_series(nc_div)

nt_s_editingLevel = to_series(nt_syn)
nt_r_editingLevel = to_series(nt_rest)
nt_d_editingLevel = to_series(nt_div)

box_colors = ['grey', 'dodgerblue', 'red']  # Syn, Rest, Div
edit_types = ["Synonymous ($\it{L}_{S}$)", "Restorative ($\it{L}_{R}$)","Diversifying ($\it{L}_{D}$)"]

species_labels = {
    'fg': r"$\it{F.\ graminearum}$",
    'nc': r"$\it{N.\ crassa}$",
    'nt': r"$\it{N.\ tetrasperma}$"
}
species_order = ['fg', 'nc', 'nt']

all_data = []
for sp in species_order:
    try:
        s_edit = eval(f"{sp}_s_editingLevel").dropna() 
        r_edit = eval(f"{sp}_r_editingLevel").dropna()
        d_edit = eval(f"{sp}_d_editingLevel").dropna() 
    except:
        s_edit = r_edit = d_edit = pd.Series(dtype=float)
    all_data.append([s_edit, r_edit, d_edit])

fig, ax = plt.subplots(figsize=(13, 6))
positions = []
box_data = []
colors = []
xticks = []
xtick_labels = []

spacing = 4
width = 0.6

for i, (sp, (s, r, d)) in enumerate(zip(species_order, all_data)):
    base = i * spacing
    box_data.extend([s, r, d])
    positions.extend([base, base + 1, base + 2])
    colors.extend(box_colors)
    xticks.append(base + 1)
    xtick_labels.append(species_labels[sp])

bp = ax.boxplot(
    box_data,
    positions=positions,
    widths=width,
    patch_artist=True,
    medianprops=dict(color='white', linewidth=1.5),
    showfliers=False
)

for patch, color in zip(bp['boxes'], colors):
    patch.set_facecolor(color)
    patch.set_linewidth(0)

ax.set_ylabel("Editing Level (%)", fontsize=22)
ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels, fontsize=22)
ax.tick_params(axis='y', labelsize=22)
ax.set_ylim(-2,85)

def get_stars(pval):
    if pval < 0.0001:
        return '****'
    elif pval < 0.001:
        return '***'
    elif pval < 0.01:
        return '**'
    elif pval < 0.05:
        return '*'
    else:
        return 'ns'

for i, (s, r, d) in enumerate(all_data):
    base = i * spacing

    if len(s) == 0 or len(r) == 0 or len(d) == 0:
        continue

    y_max = np.nanmax([s.max(), r.max(), d.max()])
    offset = 0.08* y_max  
    
    try:
        pval_r = mannwhitneyu(r, s, alternative='two-sided').pvalue
        stars_r = get_stars(pval_r)
        line_y = min(y_max + offset, 68)
        ax.plot([base, base + 1], [line_y, line_y], color='black', linewidth=1)
        text_y = line_y + (offset*0.001 if stars_r != 'ns' else offset*0.2)
        ax.text(base + 0.5, text_y, stars_r, ha='center', fontsize=22)
    except:
        pass

    try:
        pval_d = mannwhitneyu(d, s, alternative='two-sided').pvalue
        stars_d = get_stars(pval_d)
        line_y = min(y_max + 2*offset, 80)
        ax.plot([base, base + 2], [line_y, line_y], color='black', linewidth=1)
        text_y = line_y + (offset*0.08 if stars_d != 'ns' else offset*0.5)
        ax.text(base + 1, text_y, stars_d, ha='center', fontsize=22)
    except:
        pass

handles = [Patch(facecolor=c, label=l) for c, l in zip(box_colors, edit_types)]

fig.tight_layout(rect=(0, 0.12, 1, 1))  
sns.despine()

ax.legend(
    handles=handles,
    frameon=False,
    fontsize=22,
    loc='upper center',
    bbox_to_anchor=(0.5, -0.10), 
    ncol=3
)
plt.show()
```
